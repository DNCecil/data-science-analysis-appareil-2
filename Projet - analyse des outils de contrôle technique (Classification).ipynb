```python
#Créer une fonction pour la méthode du winsorize
from scipy.stats.mstats import winsorize

def winsorize_columns (df, column, lower_quantile=0.05, upper_quantile=0.95):
            lower_bound = df[column].quantile(lower_quantile)
            upper_bound = df[column].quantile(upper_quantile)
            df[column] = np.clip(df[column], lower_bound, upper_bound)
            return df
```

```python
df = winsorize_columns(df, 'humidite', lower_quantile=0.05, upper_quantile=0.95)
df = winsorize_columns(df, 'precipitation', lower_quantile=0.05, upper_quantile=0.95)
df = winsorize_columns(df, 'variation pression', lower_quantile=0.05, upper_quantile=0.95)
df = winsorize_columns(df, 'vitesse vent', lower_quantile=0.05, upper_quantile=0.95)
df.describe()
```

```python
#suppression des doublons
#creation d'une fonction de suppression des doublons
def remove_duplicates (df, keep="first"):
    
    before = df.shape[0] #nombre de ligne avant la suppression
    df_cleaned = df.drop_duplicates(keep=keep)
    after = df_cleaned.shape[0] #nombre de ligne après la suppression
    
    print(f"{before - after} doulon(s) supprimé(s).")
    return df_cleaned
```

### Classification modèles

```python
#-----------------------------
# 1. Decision Tree Classifier
#-----------------------------
dt_model = DecisionTreeClassifier(max_depth=20, random_state=50)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
```

```python
#-----------------------------
# 2. Random Forest Classifier
#-----------------------------
rf_model = RandomForestClassifier(n_estimators=1000, max_depth=20, random_state=20)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
#-------------------------------------
# 3. Support Vector Classifier (SVC)
#-------------------------------------

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svc_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svc_model.fit(X_train_scaled, y_train)
svc_pred = svc_model.predict(X_test_scaled)
```

```python
#-----------------------------
# 4. Logistic Regressor
#-----------------------------
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

### Classification models evaluations

```python
# Evaluation du Decision Tree model
dt_acc = accuracy_score(y_test, dt_pred)
dt_clr = classification_report(y_test, dt_pred)
dt_roc_curve = roc_curve(y_test, dt_pred)
print("The accuracy score is :", dt_acc)
print("The classification report is :", dt_clr)
print("The roc curve is as follows :", dt_roc_curve)
```

```python
# Evaluation du Random Forest model
rf_acc = accuracy_score(y_test, rf_pred)
rf_clr = classification_report(y_test, rf_pred)
rf_roc_curve = roc_curve(y_test, rf_pred)
print("The accuracy score is :", rf_acc)
print("The classification report is :", rf_clr)
print("The roc curve is as follows :", rf_roc_curve)
```

```python
# Evaluation du SVC model
svc_acc = accuracy_score(y_test, svc_pred)
svc_clr = classification_report(y_test, svc_pred)
svc_roc_curve = roc_curve(y_test, svc_pred)
print("The accuracy score is :", svc_acc)
print("The classification report is :", svc_clr)
print("The roc curve is as follows :", svc_roc_curve)
```

```python
# Evaluation du Logistic Regressor model
lr_acc = accuracy_score(y_test, lr_pred)
lr_clr = classification_report(y_test, lr_pred)
lr_roc_curve = roc_curve(y_test, lr_pred)
print("The accuracy score is :", lr_acc)
print("The classification report is :", lr_clr)
print("The roc curve is as follows :", lr_roc_curve)
```
